```python
def scrape_tps(noktp):
  driver.get(url)
  sleep(2)

  driver.find_element(By.ID, "__BVID__19").send_keys(noktp)
  driver.find_element(By.XPATH, "(//span[contains(text(), 'Pencarian')])[2]").click()
  sleep(2)

  soup = driver.page_source
  bsobj = BeautifulSoup(soup,'html.parser')
  name = bsobj.find('div', {'class': 'row row-1'}).p.text.replace("Nama Pemilih", "")
  NoTPS = bsobj.find('p', {'class': 'row--right'}).text.replace("TPS", "")
  Alamat = bsobj.findAll('div', {'class': 'row row-3'})
  AlamatTPS = Alamat[1].p.text.replace('Alamat TPS','')


  Other = bsobj.find('div', {'class': 'row row-3'})
  Kelurahan = Other.find('p', {'class': 'row--right'}).text.replace("Kelurahan", "")
  Kecamatan = Other.find('p', {'class': 'row--center'}).text.replace("Kecamatan", "")
  Kabupaten = Other.find('p', {'class': 'row--left'}).text.replace("Kabupaten", "")


  data = {'Nama': [name],
          'NIK': [noktp],
          'Nomor TPS': [NoTPS],
          'Alamat TPS': [AlamatTPS],
          'Kelurahan': [Kelurahan],
          'Kecamatan': [Kecamatan],
          'Kabupaten': [Kabupaten]}

  df = pd.DataFrame(data)
  return df
```
