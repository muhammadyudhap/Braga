```python
def date_indo_world(tanggal_indo=None):
    new_date = tanggal_indo.lower()
    world_date = new_date.replace('jan', 'Jan').replace('feb', 'Feb').replace('mar', 'Mar').replace('apr', 'Apr').replace('mei', 'May').replace('jun', 'Jun').replace('jul', 'Jul').replace('agt', 'Aug').replace('sep', 'Sep').replace('okt', 'Oct').replace('nov', 'Nov').replace('des', 'Dec')
    return datetime.strptime(world_date, "%d %b %Y")
```

```python
def tiketdotcom_detail(url):
    driver.get(url)
    soup = driver.page_source
    bsobj = BeautifulSoup(soup,'html.parser')
    
    Event_name = bsobj.find('h1',{'class':'HcPVsG_text HcPVsG_size_h2'}).text
    Start_date = bsobj.find('span',{'class':'SectionSummary_column_wrapper__Riju8 HcPVsG_text HcPVsG_size_b2'}).text[0:11]
    Start_date = date_indo_world(Start_date)
    
    try:
        Location = bsobj.find('span',{'class':'SectionLocation_full_address__RgtLU HcPVsG_text HcPVsG_size_b2'}).text
    except:
        Location = bsobj.find('div',{'class':'Breadcrumb_breadcrumb__a2wU1 productSlug_breadcrumb__BoXmN'}).text
        
    # Address = bsobj.find('span',{'class':'SectionLocation_street__5nfUn HcPVsG_text HcPVsG_size_b1 HcPVsG_weight_bold'}).text
    try:
        Price=bsobj.find('span',{'class':'HcPVsG_text HcPVsG_variant_disabled HcPVsG_size_h3 HcPVsG_weight_bold'}).text
    except:
        Price=bsobj.find('span',{'class':'HcPVsG_text HcPVsG_variant_alert HcPVsG_size_h3 HcPVsG_weight_bold'}).text
    
    Price = int(Price.replace('IDR','').replace('.',''))
    
    try:
        driver.execute_script("window.scrollTo(0,1000)")
        driver.find_element(By.XPATH, "//span[contains(text(), 'Deskripsi')]").click()
        sleep(1)
        soup = driver.page_source
        bsobj = BeautifulSoup(soup,'html.parser')
        Desc = bsobj.find('section',{'class':'SectionDescription_section_container__SU6j4'}).text.replace('Deskripsi','')
    except:
        Desc = None
                                 
    data = {
        'Event Name': [Event_name],
        'Start Date': [Start_date],
        'Location': [Location],
        'Price': [Price],
        'Description':[Desc],
        'Source': [url]
    }
    
    df = pd.DataFrame(data)   
    return df
    
```

```python
def tiketdotcom_detail(url):
    driver.get(url)
    soup = driver.page_source
    bsobj = BeautifulSoup(soup,'html.parser')
    
    Event_name = bsobj.find('h1',{'class':'HcPVsG_text HcPVsG_size_h2'}).text
    Start_date = bsobj.find('span',{'class':'SectionSummary_column_wrapper__Riju8 HcPVsG_text HcPVsG_size_b2'}).text
    Location = bsobj.find('span',{'class':'SectionLocation_full_address__RgtLU HcPVsG_text HcPVsG_size_b2'}).text
    Address = bsobj.find('span',{'class':'SectionLocation_street__5nfUn HcPVsG_text HcPVsG_size_b1 HcPVsG_weight_bold'}).text
    
    try:
        Price=bsobj.find('span',{'class':'HcPVsG_text HcPVsG_variant_disabled HcPVsG_size_h3 HcPVsG_weight_bold'}).text
    except:
        Price=bsobj.find('span',{'class':'HcPVsG_text HcPVsG_variant_alert HcPVsG_size_h3 HcPVsG_weight_bold'}).text
                                 
    try:
        Event_status = bsobj.find('div',{'class':'HcPVsG_text HcPVsG_variant_lowEmphasis HcPVsG_size_h3 HcPVsG_weight_bold'}).text
    except:
        Event_status = None
                                 
    data = {
        'Event Name': [Event_name],
        'Start Date': [Start_date],
        'Location': [Location],
        'Address': [Address],
        'Price': [Price],
        'Event Status': [Event_status]
    }
    
    df = pd.DataFrame(data)   
    return df
    
```
